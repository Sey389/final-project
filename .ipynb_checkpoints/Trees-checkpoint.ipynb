```python
import pandas as pd
import geopandas
%matplotlib inline
```

```python
# import dataset
stems = pd.read_csv("TreeData.csv", low_memory=False)
# filter to important info
num_stems = stems.groupby("neighborhood").sum()['stems']
```

```python
# do the merge
neighborhoods = geopandas.read_file("Neighborhoods/Neighborhoods_.shp")
stems_map = neighborhoods.merge(num_stems, how='left', left_on='hood', right_on='neighborhood')
# look at the head to confirm it merged correctly
```

```python
stems_map.plot(column='stems', # set the data to be used for coloring
                cmap = 'Purples', # choose a color palette
               edgecolor="black",        # outline the districts in white
               legend=True,              # show the legend
               legend_kwds={'label': "Number of Stems"}, # label the legend
               figsize=(15, 10),# set the size
               )
```

Conclusion: After looking at the map, it is obvious that there is a winner that stand out compared to the rest of the neighborhood. The winner of the most "Green" neighborhood belongs to squirrel hill south! This also makes sense as Schelney Park is in that neighborhood which contains a lot of trees compared to other neighborhoods.
